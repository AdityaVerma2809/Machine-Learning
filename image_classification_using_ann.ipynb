```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test)=keras.datasets.cifar10.load_data()
```

```python
x_train.shape
```

```python
x_train[0].shape
```

```python
plt.imshow(x_train[0])
```

```python
y_train[0][0]
```

```python
plt.imshow(x_train[1])
```

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
classes[y_train[1][0]]
```

```python
x_train_scaled=x_train/255 #normalising
x_test_scaled=x_test/255
```

```python
y_train_categorical = keras.utils.to_categorical(y_train,num_classes=10)
y_test_categorical = keras.utils.to_categorical(y_test,num_classes=10)
y_train_categorical[:5]
```

```python
model =keras.Sequential([
     keras.layers.Flatten(input_shape=(32,32,3)),
     keras.layers.Dense(3000,activation='relu'),
     keras.layers.Dense(1000,activation='relu'),
     keras.layers.Dense(100,activation='relu'),
     keras.layers.Dense(10, activation='sigmoid')
 ])

model.compile(optimizer='SGD',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train_scaled,y_train_categorical,epochs=50)
```

```python
plt.imshow(x_test_scaled[0])
```

```python
classes[y_test[0][0]]
```

```python
model.predict(x_test_scaled)[0]# the output shows the probability of each of the classes
```

```python
np.argmax(model.predict(x_test_scaled)[0])# gives the index of the largest element within the array
```

```python
classes[np.argmax(model.predict(x_test_scaled)[0])]
```

```python
model.evaluate(x_test_scaled,y_test_categorical)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(y_test,np.argmax(model.predict(x_test_scaled),axis=1))
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Header Parameter in pd.read_csv



*   Default Behavior (header=0): By default, read_csv() assumes the first row (index 0) of the CSV file contains the column headers
*   No Header Row (header=None): If your CSV file does not contain a header row, you can set header=None. Pandas will then automatically assign integer column names (0, 1, 2, ...).





```python
df=pd.read_csv('sonar_dataset.csv', header=None)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[60].value_counts()
```

```python
x=df.drop(columns=60, axis=1)
y=df[60]
```

```python
y=pd.get_dummies(y, drop_first=True)
y.sample(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.25, random_state=1)
```

```python
x_train.shape, x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train, y_train, epochs=100, batch_size=8)
```

model is perfect fit for the training dataset but lets see the evalution on the test dataset

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test).reshape(-1)
print(y_pred[:10])

y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
model1=keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

model1.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model1.fit(x_train, y_train, epochs=100, batch_size=8)
```

```python
model1.evaluate(x_test,y_test)
```

```python
y_pred=model1.predict(x_test).reshape(-1)
y_pred=np.round(y_pred)

print(classification_report(y_test, y_pred))
```
